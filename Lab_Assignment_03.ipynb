# **TASK - 1**

```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')

sentence = "Students are Learning Natural Learning Processing"
tokens = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(tokens)
print(pos_tags)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("Students are Learning Natural Language Processing")
for token in doc:
    print(token.text, token.pos_)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying a Startup in India.")
for token in doc:
  print(token.text, token.pos_)
```

```python
import spacy
from collections import Counter
nlp = spacy.load("en_core_web_sm")
text = "Loving the new AI features üòç #AI #MachineLearning"
doc = nlp(text)
nouns = []
verbs = []
for token in doc:
  if token.pos_ in ["NOUN", "PROPN"]:
    nouns.append(token.text)
  elif token.pos_ == "VERB":
    verbs.append(token.text)
noun_freq = Counter(nouns)
verb_freq = Counter(verbs)
print("Noun Frequency:", noun_freq)
print("Verb Frequency:", verb_freq)
```

```python
words = """ Academic writing plays a crucial role in the dissemination of knowledge.
Researchers analyze data, develop theories, and present arguments in a structured manner.
The study demonstrates how linguistic patterns influence clarity and coherence.
Effective communication enhances understanding and supports critical evaluation.
"""

import nltk
nltk.download('punkt_tab')
nltk.download('average_perceptron_tagger_eng')
words_tokens = nltk.word_tokenize(words)
words_pos_tags = nltk.pos_tag(words_tokens)
print(words_pos_tags)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("""Academic writing plays a crucial role in the dissemination of knowledge.
Researchers analyze data, develop theories, and present arguments in a structured manner.
The study demonstrates how linguistic patterns influence clarity and coherence.
Effective communication enhances understanding and supports critical evaluation.
""")
for words_tokens in doc:
  print(words_tokens.text, words_tokens.pos_)
```
