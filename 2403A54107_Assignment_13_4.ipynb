```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=categories))
```

# ***Task - 2:***

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.datasets import fetch_20newsgroups

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
```

```python
categories = ['sci.space', 'rec.sport.hockey']
data = fetch_20newsgroups(subset='all', categories=categories)

texts = data.data
labels = data.target

for i in range(3):
    print("Text:", texts[i][:200])
    print("Label:", labels[i])
    print("------")
```

```python
import re

def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text

processed_texts = [preprocess(t) for t in texts]
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(processed_texts)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.2, random_state=42
)
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Predictions:", y_pred)
print("Actual:", y_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import joblib

joblib.dump(model, 'naive_bayes_model_news.pkl')

joblib.dump(vectorizer, 'vectorizer_news.pkl')
```

```python
loaded_model = joblib.load('naive_bayes_model_news.pkl')

loaded_vectorizer = joblib.load('vectorizer_news.pkl')
```

```python
test_samples = [
    "NASA launched a new satellite into space",
    "The hockey team won the championship game",
    "3D rendering and computer graphics techniques improved"
]

test_vec = loaded_vectorizer.transform(test_samples)

predictions = loaded_model.predict(test_vec)

for text, pred in zip(test_samples, predictions):
    print("\nText:", text)
    print("Predicted Category:", data.target_names[pred])
```
